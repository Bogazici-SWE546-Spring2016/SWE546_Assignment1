```python
%matplotlib inline

import scipy as sc
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt

df_arac = pd.read_csv(u'data/arac.csv',sep=';')
BaseYear = 1966
degree = 2

x = np.matrix(df_arac.Year[0:]).T-BaseYear
xx = np.matrix(np.linspace(-2,55)).T
A2 = np.hstack((np.power(xx,i) for i in range(degree+1)))

xTarget = np.mat('[2016;2017;2018;2019;2020]')-BaseYear
ATarget = np.hstack((np.power(xTarget,i) for i in range(degree+1)))

# replace empty values with 0. Is it the best option?????
df_arac = df_arac.replace(np.nan,0, regex=True)

# iterate the columns of the dataset except year column 
for column in df_arac.columns[1:]:
    
    y = np.matrix(df_arac[column][0:]).T/1000000
    A = np.hstack((np.power(x,i) for i in range(degree+1)))
    
    # Solve the least squares problem
    w_ls = (A.T*A).I*A.T*y
    #w_ls,E,rank,sigma = np.linalg.lstsq(A, y)

    fTarget = ATarget * w_ls
    f = A2*w_ls

    plt.plot(x+BaseYear, y, 'o-')
    plt.plot(xx+BaseYear, f, 'r')
    plt.xlabel('Years')
    plt.ylabel('Number of '+column)
    plt.show()

    for i in range(len(fTarget)):
        print ("Number of %s in %s: %s" % (column,2016+i,fTarget[i]))
```
