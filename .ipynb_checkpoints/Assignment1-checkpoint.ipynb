```python
%matplotlib inline

import scipy as sc
import scipy.stats as scst
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pylab as plt

df_arac = pd.read_csv(u'data/arac.csv',sep=';')
df_arac
```

```python
BaseYear = 1966

x1 = np.matrix(df_arac.Year[0:]).T-BaseYear

y_all = np.matrix(df_arac.All[0:]).T/1000000.0
y_car = np.matrix(df_arac.Car[0:]).T/1000000.0
y_minibus = np.matrix(df_arac.Minibus[0:]).T/1000000.0
y_bus = np.matrix(df_arac.Bus[0:]).T/1000000.0
y_smallTruck = np.matrix(df_arac.SmallTruck[0:]).T/1000000.0
y_truck = np.matrix(df_arac.Truck[0:]).T/1000000.0
y_motorcycle = np.matrix(df_arac.Motorcycle[0:]).T/1000000.0
y_specialVehicles = np.matrix(df_arac.SpecialVehicles[0:]).T/1000000.0
y_machinery = np.matrix(df_arac.Machinery[0:]).T/1000000.0
y_tractor = np.matrix(df_arac.Tractor[0:]).T/1000000.0
#y = np.matrix(df_arac.Minibus[0:-1:5]).T/1000000.0

plt.plot(x1+BaseYear, y_car, 'o-')
plt.xlabel('Year')
plt.ylabel('Number of Cars (Millions)')

plt.show()
```

```python
#TargetYear = np.arange(1966,2016,1)
t = np.arange(0,50,1)
BaseYear = 1966

def square(list):
    return [i ** 2 for i in list]

def fitit(a,b):
    coeffs = np.polyfit(t,a,deg=b)
    t_x=np.squeeze(np.asarray(coeffs.T))
    pf = np.polyval(np.poly1d(t_x), t)
    plt.plot(x1+BaseYear, a , 'o-')
    plt.plot(x1+BaseYear, pf , 'r')
    plt.xlabel('Year')
    plt.ylabel('Number of Cars (Millions)')
    plt.show()
    print('2016 : %s' % np.polyval(np.poly1d(t_x), 50))
    print('2017 : %s' % np.polyval(np.poly1d(t_x), 51))
    print('2018 : %s' % np.polyval(np.poly1d(t_x), 52))
    print('2019 : %s' % np.polyval(np.poly1d(t_x), 53))
    print('2020 : %s' % np.polyval(np.poly1d(t_x), 54))
    a = np.squeeze(np.asarray(a))
    print('Maximum Absolute Error %.1f' % max(a-pf))
    print('Total Square Diffrence %.1f' % sum(square(a - pf)))

def fititx(a):
    print('K = 1')
    fitit(a,1)    
    print('K = 2')
    fitit(a,2)    
    print('K = 3')
    fitit(a,3)    
    print('K = 4')
    fitit(a,4)    
    print('K = 5')
    fitit(a,5)    
    print('K = 6')
    fitit(a,6)    
    print('K = 7')
    fitit(a,7)    
    print('K = 8')
    fitit(a,8)    
    print('K = 9')
    fitit(a,9)    
    print('K = 10')
    fitit(a,10)    

print('---------------------------------------------')
print('All Vehicles')
print('---------------------------------------------')
fititx(y_all)
print('---------------------------------------------')
print('Cars')
print('---------------------------------------------')
fititx(y_car)
print('---------------------------------------------')
print('Minibus')
print('---------------------------------------------')
fititx(y_minibus)
print('---------------------------------------------')
print('Bus')
print('---------------------------------------------')
fititx(y_bus)
print('---------------------------------------------')
print('Small Truck')
print('---------------------------------------------')
fititx(y_smallTruck)
print('---------------------------------------------')
print('Truck')
print('---------------------------------------------')
fititx(y_truck)
print('---------------------------------------------')
print('Motorcycle')
print('---------------------------------------------')
fititx(y_motorcycle)
print('---------------------------------------------')
print('Special Vehicles')
print('---------------------------------------------')
fititx(y_specialVehicles)
print('---------------------------------------------')
print('Machinery')
print('---------------------------------------------')
fititx(y_machinery)
print('---------------------------------------------')
print('Tractor')
print('---------------------------------------------')
fititx(y_tractor)
```
